```python
import pandas as pd

# Lê o CSV e converte para uma lista de dicionários
users = pd.read_csv('SDW2023.csv').to_dict(orient='records')

# Garante a estrutura esperada para a etapa de Transformação
for user in users:
    user['news'] = []

print(users)
```

```python
import random

# 1.Banco de mensagens de marketing
mensagens_marketing = [
    "Invista hoje e garanta seu futuro com nossos CDBs!",
    "Descubra as vantagens do novo cartão de crédito sem anuidade.",
    "Organize suas finanças com nossa nova ferramenta de gestão.",
    "Aproveite as taxas reduzidas para financiamento imobiliário este mês.",
    "Diversifique seus investimentos com nossos fundos multimercado."
]

# 2. Etapa de Transformação: associar uma mensagem a cada cliente
for user in users:
    # Sorteia uma mensagem aleatória da nossa lista
    mensagem_sorteada = random.choice(mensagens_marketing)

    # Salva a mensagem dentro do espaço 'news' do cliente, no formato exigido
    user['news'].append({
        "description": mensagem_sorteada
    })

# 3. Visualizar o resultado final transformado
print(users)
```

```python
# 1. Prepara os dados para o formato de tabela
dados_finais = []
for user in users:
  dados_finais.append({
      "UserId": user['UserID'],
      "Mensagem_Marketing": user['news'][0]['description']
  })

# 2. Transforma de volta para o formato do pandas
df_final = pd.DataFrame(dados_finais)

# 3. Etapa de Load: Salva em um novo arquivo CSV
df_final.to_csv('SDW2023_Resultados.csv', index=False)

# Configuração para remover o limite de caracteres da coluna na exibição
pd.set_option('display.max_colwidth', None)

# Imprime a nossa tabela
print("Etapa de Load concluída com sucesso!")
print(df_final)
```
